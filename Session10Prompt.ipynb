```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_uniform = np.random.uniform(0, 1, 1000)
```

```python
plt.hist(data_uniform, bins=100, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Uniform Random Numbers")
plt.grid(True)
plt.tight_layout()
plt.savefig("uniform_histogram.pdf")
plt.show()
```

```python
data_beta = np.random.beta(a=2, b=5, size=1000)
```

```python
plt.hist(data_beta, bins=100, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Beta(2,5) Random Numbers")
plt.grid(True)
plt.tight_layout()
plt.savefig("beta_histogram.pdf")
plt.show()
```
